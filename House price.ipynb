### The purpose of the analysis: Analyze / summarize the price of houses in a community and understand why price varies. Use Seaborn library wherever it makes sense.

## libs

```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Summarize house prices in the community. Add an appropriate chart.

```python
house_prices_file = 'house_prices.xlsx'
house_prices = pd.read_excel(house_prices_file, header=0, index_col=False, keep_default_na=True)
house_prices.describe()
```

```python
house_prices['price'].plot().axes.get_xaxis().set_visible(False)
plt.ylabel('Prices')
plt.title("Prices in the Community");
```

## Create a visual to show the variation in house prices, as the number of rooms in the house changes. Write down your observations

```python
sns.boxplot(data=house_prices,x='rooms',y='price')
plt.title("Variation in house prices / rooms");
```

We can see that the variation of prices is important for apartments of 8 rooms but for 9, 10 and 11 prices are do not vary a lot.

## Explore the relation between price and the other attributes of the house. Use charts, tables etc. Feel free to use any number of charts and tables.

```python
# Price and Size_SquareFeet
sns.jointplot(data=house_prices, x= "size_sqft", y="price", kind= "hist");
```

```python
# Price and Acres
sns.jointplot(data=house_prices, x= "acres", y="price", kind= "hist");
```

```python
# Price and Bathrooms
sns.jointplot(data=house_prices, x= "baths", y="price", kind= "hist");
```

```python
# Price and Age
sns.jointplot(data=house_prices, x= "age", y="price", kind= "hist");
```

```python
# Price and Garages
sns.jointplot(data=house_prices, x= "garages", y="price", kind= "hist");
```

```python
house_prices_correlation = house_prices.corr()
house_prices_correlation
```

```python
plt.figure(figsize = [18,10])
sns.heatmap(data= house_prices_correlation, annot = True, cmap="YlGnBu")
plt.title("Correlation Matrix");
```
